```python
from IPython.display import clear_output
import sys
#import ipywidgets as widgets
#from IPython.display import display
board=[0,0,0,0,0,0,0,0,0]
def show_board(board):
    clear_output()
    count=0
    for i in range(9):
        if i%3==0:
            print()
            if i>2:
                print('-----------')
        if i%3>0:
            print('|', end='')
        if board[i]==0:
            #wig=widgets.Checkbox(value=False, disabled=False)
            #display(wig)
            sym='   '
        elif board[i]==1:
            sym=' X '
            count+=1
        else:
            sym=' O '
            count+=1
        print(sym, end='')
    print()
    return count
        
def check_win(board):
    for i in range(3):
        if board[0+i*3]+board[1+i*3]+board[2+i*3]==3\
        or board[i]+board[3+i]+board[6+i]==3:
               return 10
    if board[0]+board[4]+board[8]==3\
    or board[6]+board[4]+board[2]==3:
        return 10
    
    for i in range(3):
        if board[0+i*3]+board[1+i*3]+board[2+i*3]==-3\
        or board[i]+board[3+i]+board[6+i]==-3:
               return -10
    if board[0]+board[4]+board[8]==-3\
    or board[6]+board[4]+board[2]==-3:
        return -10
    
    return 0
    
def minimax(a_board, step, side):
    #if step>9:
    #    return 0
    global bestmove
    if side==1:
        maxscore=-1000
    else:
        maxscore=1000
    for i in range(9):
        if a_board[i]!=0:
            continue
        board=a_board.copy()
        board[i]=side
        score=check_win(board)
        if (score==0):
            score=minimax(board,step+1,-side)
        if side==1:
            if score-step>maxscore:
                maxscore=score-step
                if step==0:
                    bestmove=i
        else:
            if score+step<maxscore:
                maxscore=score+step
                if step==0:
                    bestmove=i
        
    if abs(maxscore)==1000:
        return 0
    else:
        return maxscore
if __name__ == '__main__':
    total=show_board(board)

    while(total<9):
        bestmove=-1
        sys.stdout.flush()
        move=input("input your move as:\n0 1 2\n3 4 5\n6 7 8\n")
        try:
            move=int(move)
        except:
            print('wrong input!')
            continue
        if board[move]!=0:
            print('Not legal move!')
            continue
        board[move]=1
        total = show_board(board)
        win=check_win(board)
        if win==10:
            print('you win!')
            break
    
        if total<2 and board[4]==0:
            bestmove=4
        else:
            minimax(board,0,-1)
        board[bestmove]=-1
        win=check_win(board)
    
        total = show_board(board)
        if win==-10:
            print('you lose!')
            break
    if total>=9:
        print('It is a draw!')
    
```
